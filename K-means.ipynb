```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris=pd.read_csv('iris.csv')
```

```python
iris
```

```python
Iris_data=iris.drop(['Id','Species'],axis=1)
```

```python
Iris_data
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for k in  range(1,15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(Iris_data)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(20,8))
plt.title("WCSS /K Chart", fontsize=18)
plt.plot(range(1,15),wcss,"-o")
plt.grid(True)
plt.xlabel("Amount of Clusters",fontsize=14)
plt.ylabel("Inertia",fontsize=14)
plt.xticks(range(1,20))
plt.tight_layout()
plt.show()
```

```python
kmeans_predict=kmeans.predict(Iris_data)
```

```python
kmeans_predict
```

```python
Iris_data=pd.DataFrame(Iris_data)
```

```python
Iris_data["Label"]=kmeans_predict
```

```python
Iris_data
```

```python
plt.figure(figsize=(30,5))     
plt.subplot(1,5,5)
plt.title("Original Labels",fontsize=16)
plt.xlabel("PetalLengthCm")
plt.scatter(Iris_data.SepalLengthCm[Iris_data.Label == "Lily"],Iris_data.SepalWidthCm[Iris_data.Label == "Lily"])
plt.scatter(Iris_data.SepalLengthCm[Iris_data.Label == "Lotus"],Iris_data.SepalWidthCm[Iris_data.Label == "Lotus"])
plt.scatter(Iris_data.SepalLengthCm[Iris_data.Label == "Hibiscus"],Iris_data.SepalWidthCm[Iris_data.Label == "Hibiscus"])
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
sns.pairplot(data=Iris_data,hue="Label",palette="Set1")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
x = silhouette_score
```

```python
silhouette_score(Iris_data,kmeans_predict)
```

```python
X=Iris_data.iloc[:,:-1]
Y=Iris_data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
classification_report(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
knn.score(x_test,y_pred)
```
